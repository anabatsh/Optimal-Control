```python
N = 30
time_step = 0.05
x_0 = 0.8
x_ref = 0.7
```

```python
def f(x, u, i):
    return x - i

def h(x, u, i):
    return x - x

def F1(x, u, i):
    return x - x_ref

def F(x, u, i):
    return 0.5 * (F1(x, u, i) ** 2)
```

```python
history_nlp = run_optimal_control(x_0, x_ref, N, time_step, f, h, F, integer=False, name='NLP')
show_results(history_nlp)
```

```python
x_nlp = history_nlp['x']
u_nlp = history_nlp['u']
i_nlp = history_nlp['i']
obj_nlp = history_nlp['obj_value']

f_L  = lambda x, x0: linear(f, x, x0)
h_L  = lambda x, x0: linear(h, x, x0)
F_L  = lambda x, x0: linear(F, x, x0)
F_QP = lambda x, x0: gauss_newton(F, F1, x, x0)
F_GN = lambda x, x0: F(*x) #F_QP(x, x0) #- obj_nlp
```

```python
history_dist = run_optimal_control_dist(x_0, x_ref, x_nlp, u_nlp, i_nlp, N, 
                                        time_step, f_L, h_L, F_GN, F, F_L, 'DIST')
show_results(history_dist)
```

```python
y_x0  = history_dist['F0']
y_x   = history_dist['F']
y_x_l = history_dist['F_L']
y_x_g = history_dist['F_GN']
```

```python
plt.figure(figsize=(15, 5))

time = range(N+1)
plt.plot(time, y_x0, '-o', markersize=4, c='red', label='known points')
plt.plot(time, y_x, '-o', markersize=4, c='black', label='true points')
plt.plot(time, y_x_l, '-o', markersize=4, c='green', label='linear approx')
plt.plot(time, y_x_g, '-o', markersize=4, c='blue', label='gauss-newton approx')

plt.xticks(time)
plt.grid()

plt.legend()

plt.show()
```
